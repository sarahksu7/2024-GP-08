 Train and evaluate a Multi-Output Linear Regression model to predict UPDRS scores using K-Fold Cross-Validation.

```python
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Load datasets with full file paths
files = [
    r'C:\Users\sarah\Desktop\saragp\dataset_cleaned\t_plus_0.csv',
    r'C:\Users\sarah\Desktop\saragp\dataset_cleaned\t_plus_6.csv',
    r'C:\Users\sarah\Desktop\saragp\dataset_cleaned\t_plus_12.csv',
    r'C:\Users\sarah\Desktop\saragp\dataset_cleaned\t_plus_24.csv'
]
dataframes = [pd.read_csv(file) for file in files]
data = pd.concat(dataframes, ignore_index=True)

# Define features (X) and targets (y)
X = data.iloc[:, 10:]  # Features starting from column 10
y = data[['updrs_1', 'updrs_2', 'updrs_3', 'updrs_4']]  # Targets

# Initialize Multi-Output Linear Regression
lr = LinearRegression()
multi_output_lr = MultiOutputRegressor(lr)

# Function to perform K-Fold Cross-Validation
def perform_kfold_cv_separate(X, y, k):
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    results = {col: {"r2": [], "mse": []} for col in y.columns}  # Initialize results

    for train_index, test_index in kf.split(X):
        X_train_cv, X_test_cv = X.iloc[train_index], X.iloc[test_index]
        y_train_cv, y_test_cv = y.iloc[train_index], y.iloc[test_index]

        multi_output_lr.fit(X_train_cv, y_train_cv)
        y_pred_cv = multi_output_lr.predict(X_test_cv)

        # Calculate metrics for each target separately
        for i, col in enumerate(y.columns):
            cv_r2 = r2_score(y_test_cv[col], y_pred_cv[:, i])
            cv_mse = mean_squared_error(y_test_cv[col], y_pred_cv[:, i])
            results[col]["r2"].append(cv_r2)
            results[col]["mse"].append(cv_mse)

    # Print results for each target
    print(f"\nK-Fold Cross-Validation with {k} Folds:")
    for col in y.columns:
        print(f"\nTarget: {col}")
        print(f"Average R2 Score: {np.mean(results[col]['r2']):.4f}")
        print(f"Average Mean Squared Error: {np.mean(results[col]['mse']):.4f}")
        print(f"R2 Scores for Each Fold: {results[col]['r2']}")
        print(f"MSE Scores for Each Fold: {results[col]['mse']}")

# Perform 5-Fold Cross-Validation
perform_kfold_cv_separate(X, y, k=5)

# Perform 10-Fold Cross-Validation
perform_kfold_cv_separate(X, y, k=10)

```
